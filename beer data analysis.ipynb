## Feature selection and train test split

```python
x = d1.iloc[:, 1:12].drop(['beer/beerId','beer/brewerId','review/overall'], axis=1)
x
```

```python
x.shape
```

```python
from sklearn import preprocessing                 
encoder = preprocessing.LabelEncoder() 
x['beer/name'] = encoder.fit_transform(x['beer/name'])
x['beer/style'] = encoder.fit_transform(x['beer/style'])
x['review/text'] = encoder.fit_transform(x['review/text'])
x
```

```python
y = d1.loc[0: ,['review/overall']]
y
```

```python
y.shape
```

```python
x_train = x.iloc[0:7000, :]
x_train
```

```python
y_train = y.iloc[0:7000]
y_train
```

```python
y_train.shape
```

```python
x_test = x.iloc[7000:, :]
x_test
```

```python
y_test = y.iloc[7000:]
y_test
```

```python
y_test.shape
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

## Model Selection

```python
from sklearn import linear_model 
```

```python
m_lr = linear_model.LinearRegression()
```
